```python
# import dependency
import matplotlib.pyplot as plt
```

```python
# Set the x axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
#plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = 'Boston')
# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare By Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
fig, ax = plt.subplots()
# Create the plot and add a label for the legend.
ax.plot(x_axis, y_axis, marker = "D", color = "green", linewidth = 2, label = "Boston")
# Create the labels for the x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title
ax.set_title("PyBer Fare By Month")
# Add a grid
ax.grid()
# Add the legend
ax.legend()
ax.plot(x_axis, y_axis)
```

```python
# Bar graph
## Create the plot
plt.bar(x_axis, y_axis, color = "green", label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Create a title
plt.title("PyBer Fare By Month")
# create the legend
plt.legend()
```

```python
# scatter plots with plt.plot
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# run the plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# scatter plots with plt.scatter
plt.scatter(x_axis, y_axis)
```

```python
# scatter plot skill drill
plt.scatter(y_axis, x_axis, color = "red", label = 'Chicago')
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare By Month")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# bubble chart
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# make the dots larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# using "List Comprehension" for refactor bubble chart code
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# scatter plot with object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# bubble chart with object-oriented approach
fig,ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# bubble chart skill drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color = "cyan", edgecolors = "black", linewidth = 2, alpha = .8, label = "Boston")
# Create the labels for the x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# Create a title
ax.set_title("PyBer Fare By Month")
# create a limit for the x axis values
ax.set_xlim(0, 50)
# Add the legend
ax.legend()
plt.gca().invert_yaxis()
```

```python
# pie chart MATLAB
plt.pie(y_axis, labels = x_axis)
```

```python
#make it better
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8))
explode_vales = [0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
plt.pie(y_axis, explode = explode_vales, labels = x_axis, colors = colors, autopct = '%.1f%%')
```

```python
# pie chart skill drill
fig, ax = plt.subplots()

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = [0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
ax.pie(y_axis, explode = explode_values, colors = colors, labels = x_axis, startangle = 260, autopct = '%.1f%%', shadow = True)

fig.set_size_inches(8,8)

plt.gca().invert_yaxis()
```
